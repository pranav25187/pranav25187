```python
url = "https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv"
df = pd.read_csv(url)
```

```python
df = df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['continent'], inplace=True)
```

```python
df.fillna(0, inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'], format="%Y-%m-%d", errors='coerce')

```

```python
df['month'] = df['date'].dt.month
```

```python
df_groupby = df.groupby('continent').max().reset_index()
df_groupby
```

```python
df_groupby.to_csv("df_groupby.csv", index=False)
```

```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
df_groupby.head()
```

```python
sns.histplot(df['gdp_per_capita'], bins=30, kde=True)
plt.title("Distribution of GDP per Capita")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['total_cases'], y=df['gdp_per_capita'])
plt.xlabel("Total Cases")
plt.ylabel("GDP per Capita")
plt.title("Scatter Plot of Total Cases vs GDP per Capita")
plt.show()
```

```python
sns.pairplot(df_groupby)
plt.show()
```

```python
sns.catplot(data=df_groupby, x='continent', y='total_cases', kind='bar', height=5, aspect=2)
plt.title("Total Cases per Continent")
plt.show()
```
